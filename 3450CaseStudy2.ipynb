```r
#simple linear regression for maximum temperature (for discussion)
max_tempmodel <- lm(tmax ~ as.numeric(Year))
summary(max_tempmodel)
```

```r
#simple linear regression for minimum temperature
tempmodel <- lm(tmin ~ as.numeric(Year))
summary(tempmodel)
```

###Regression model equation for minimum termperature
$$y=-124.86094+0.08361x $$


**What would a slope of zero imply about the relationship between average minimum temperature and year?**

A slope of zero would imply that there is no linear relationship between year and average minimum temperature. This would indicate that the average minimum temperature remains constant over time and does not increase or decrease as the years progress.

**Are average minimum temperature and year related?**

The null hypothesis of the linear regression is that the input doesn't affect the output, and the alternative hypothesis is that the input *does* affect the output. Since the p-value of the slope is 7.2e-16, there is strong evidence that the year has a positive correlation with the average min temperature.

###Regression Assumptions
###Variance

```r
#Constant Variance
plot(Year, resid(lm(tmin ~ as.numeric(Year))),
     ylab="Residuals", xlab="year",
     main="Linear Model Residuals for Temp and Year")
abline(0, 0) #there may be a pattern here, may need to consider accounting for year-to-year lack of independence
```

The residual plot does not show a strong pattern or funnel shape, suggesting that the assumption of constant variance is reasonably satisfied.

##Normality

```r
#Normality of residuals
hist(resid(lm(tmin ~ as.numeric(Year))))
shapiro.test(resid(lm(tmin ~ as.numeric(Year))))
```

##Methods: Exploring average minimum temperature with a log transformation.

```r
logTmin <- log(tmin)
logLm <- lm(logTmin ~ as.numeric(Year))
```

```r
#scatter plot of log of average minimum temperature and year with regression line

plot(Year, logTmin, ylab = "log of average minimum temp (F)", xlab = "Year", pch = 20, cex = 1) # does it look linear?

abline(logLm)
```

```r
summary(logLm)
```

With a slope p-value very close to 0, we have convincing evidence that we should reject the null hypothesis that the input does not affect the output.

```r
slope     <- logLm$coef[2]
slopeConf <- confint(logLm,2)
```

```r
exp(slope)
```

```r
100*(exp(slope)-1)   # Back-transform estimated slope
100*(exp(slopeConf)-1) # Back-transform confidence interval
```

### Interpretation
Each increase in year is associated with a 0.2% increase in the median minimum temperature. We are 95% confident that the true median increase is between 0.16% and 0.24%
### Regression model equation
$$log(Tmin) = 0.002035 * Year - 0.325313$$
$$Median\ Tmin = e^{0.002035 * Year} * e^{-0.325313}$$
### How this model differs from linear model
The linear model will tell us if there is an additive effect on the mean of Tmin, while this log transformed model will tell us if there is a multiplicative effect on the median of Tmin.

```r
#plotting the log and linear model on the same graph

plot(Year, tmin, pch=20, ylab="Tmin", xlab="Year")

# linear model
abline(tempmodel, col="blue")

# log model (back-transformed)
curve(exp(coef(logLm)[1] + coef(logLm)[2]*x),
      add=TRUE, col="red")
```

##Methods: Comparing average minimum temperatures via ANOVA.

```r
#divide the data into 3 equal groups

SLCdata$group <- cut(as.numeric(SLCdata$Year),
                     breaks = quantile(as.numeric(SLCdata$Year),
                                       probs = c(0, 1/3, 2/3, 1)),
                     include.lowest = TRUE,
                     labels = c("Early", "Middle", "Late"))
```

```r
#report the data ranges for each group

tapply(as.numeric(SLCdata$Year), SLCdata$group, range)
```

```r
#box plots of the data

boxplot(tmin ~ group, data = SLCdata,
        xlab = "Time Period",
        ylab = "Average Minimum Temperature (F)",
        main = "Tmin by Time Period")
```

```r
#ANOVA normality assumption
model <- aov(tmin ~ group, data = SLCdata)
res <- residuals(model)
shapiro.test(res)
qqnorm(res)
qqline(res)
```

A p-value of .18 indicates that the residuals are normal.

```r
#ANOVA constant variance assumption
install.packages("lawstat")
library(lawstat)

levene.test(SLCdata$tmin, SLCdata$group)
plot(model, which = 1)
```

A p-value of .55 indicates each group has equal variance.

Independence assumption: Tmin measurements don't affect other measurements within the same group or within other groups.

```r
#ANOVA test
model <- aov(tmin ~ group, data = SLCdata)
summary(model)
```

The p-value is close to zero, so there is convincing evidence that at least one of the group means is different than another's.

```r
#Post hoc test
TukeyHSD(model)
```

Each comparison has a significant p-value, and zero is not in any of their confidence intervals. Therefore, we conclude that each group has a different mean than any other group.
